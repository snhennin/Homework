## Our project majority involves looking at publication bias of the comprehensive database that consists with 225 major depressive disorder clinical data studies with structural imaging of brain measures. In this meta-analysis study, researchers compared healthy brain structures with MDD brain structures. The publication bias mainly focuses on the overall results of studies findings in the Caudate, Putamen, Globus Pallidus, Thalamus, Temporal, and Hippocampus. Having a generalized look at the publication bias helps us determine the accuracy and precision of this overall database.

### Much of the evidence for the role of specific brain areas in psychiatry comes from "brain imaging," which involves various ways of looking at the brain. Some technologies like PET imaging and functional MRI can measure the activity of the brain either at rest or while a person does certain tasks. Other technologies, like traditional MRI, measure the brain's structureâ€”its size and shape. 
### Compared with the structure of a healthy brain, MDD was associated with lateral ventricle enlargement; larger cerebrospinal fluid volume; and smaller volumes of the basal ganglia, thalamus, hippocampus, frontal lobe, orbitofrontal cortex, and gyrus rectus. Patients during depressive episodes had significantly smaller hippocampal volume than patients during remission. This can be used in the future to aid diagnoses and further lead to personalizing treatment based off predictions of antidepressant responsiveness in the brain.

### Using the mean ratio and the standard error, I created a funnel plot in order to determine if there was a publication bias with any of our sources. In a funnel plot, you are looking for a triangle shape on an inverted y-axis. The reason we are expecting a funnel shape is because it is more likely for there to be a large amount of less precise data compared to the very precise data.

```python
#Loading the data from the excel spreadsheet
wb = xlrd.open_workbook('Project Data Collection.xlsx', on_demand = True)
ws = workbook.sheet_by_name('Comparisons')

#Plotting the data from the caudate section of the brain
x1 = caudate.loc[:,'Mean Ratio']
y1 = caudate.loc[:,'Standard Error']

#Plotting the data from the putamen section of the brain
x2 = putamen.loc[:,'Mean Ratio']
y2 = putamen.loc[:,'Standard Error']

#Plotting the data from the globus pallidus section of the brain
x3 = globus_pallidus.loc[:,'Mean Ratio']
y3 = globus_pallidus.loc[:,'Standard Error']

#Plotting the data from the thalamus section of the brain
x4 = thalamus.loc[:,'Mean Ratio']
y4 = thalamus.loc[:,'Standard Error']

#Plotting the data from the temporal section of the brain
x5 = temporal.loc[:,'Mean Ratio']
y5 = temporal.loc[:,'Standard Error']

#Plotting the data from the hippocampus section of the brain
x6 = hippo.loc[:,'Mean Ratio']
y6 = hippo.loc[:,'Standard Error']


#Create the funnel tringle
plt.plot([0.96,0.96],[0,0.1], 'k')
plt.plot([0.6,0.96],[0.1,0], 'k')
plt.plot([0.96,1.32],[0,0.1], 'k')

#Plot the brain sections
plt.scatter(x1,y1, color="red")
plt.scatter(x2,y2, color="darkorange")
plt.scatter(x3,y3, color="gold")
plt.scatter(x4,y4, color="limegreen")
plt.scatter(x5,y5, color="dodgerblue")
plt.scatter(x6,y6, color="orchid")

#Clean up the plot
plt.ylim(0,0.06)
plt.xlabel('Study Results', fontsize=12)
plt.ylabel('Less precision <--> More precision', fontsize=12)
plt.title('Publication Bias', fontsize=18)

plt.legend(['Avg. Mean Ratio','Funnel Boundary','Funnel Boundary','Caudate','Putamen','Globus Pallidus','Thalamus','Temporal','Hippocampus'], loc='upper right')
plt.gca().invert_yaxis()

fig1 = plt.gcf()
plt.show()
plt.draw
fig1.savefig('publication_bias.png', dpi=100)
```
